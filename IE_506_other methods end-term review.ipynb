# text preprocessing

```python
##text preprocessing
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['Text'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
## TFidf Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_v=TfidfVectorizer(max_features=5000,ngram_range=(1,3))
X=tfidf_v.fit_transform(corpus).toarray()
```

```python
""" Random Forest Classifier Model """

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(
    n_estimators = 100,  # Use 100 trees
    n_jobs = -1 # Use all CPU cores
)

rf.fit(X, s)

""" Store the scores assigned by this approach """

results = pd.DataFrame({
    'truth'      : y_orig,   # The true labels
    'label'      : s,        # The labels that model sees
    'output_std' : rf.predict_proba(X)[:,1]   # Probability Scores
}, columns = ['truth', 'label', 'output_std'])
```

# Bagging Model of PU

1.Create a training set by combining all positive data points with a random sample from the unlabeled points, with replacement.

2.Build a classifier from this "bootstrap" sample, treating positive and unlabeled data points as positives and negatives,         respectively.

3.Apply the classifier to whatever unlabeled data points were not included in the random sample – hereafter called OOB ("out of   bag") points – and record their scores.

4.Repeat the three steps above many times and finally assign to each point the average of the OOB scores it has received.

```python
X = pd.DataFrame(X)
```

```python
X['s']=s
```

```python
X
```

```python
from sklearn.tree import DecisionTreeClassifier
estimator = DecisionTreeClassifier(random_state=42)

""" Indices of positive and unlabeled data points """
iP = s[s > 0].index
iU = s[s <= 0].index

""" For each data point, keep track of how many times it has been OOB """
num_oob = pd.DataFrame(np.zeros(shape = s.shape), index = s.index)
sum_oob = pd.DataFrame(np.zeros(shape = s.shape), index = s.index)

n_estimators = 1000
for _ in range(n_estimators):
    # Get a bootstrap sample of unlabeled points for this round
    ib = np.random.choice(iU, replace=True, size = len(iP))

    # Find the OOB data points for this round
    i_oob = list(set(iU) - set(ib))

    # Get the training data (ALL positives and the bootstrap sample of unlabeled points) and build the tree
    Xb = pd.concat([X[s > 0],X.loc[ib]])
    yb = pd.concat([s[s > 0], s.loc[ib]])
    estimator.fit(Xb, yb)
    
    # Record the OOB scores from this round
    sum_oob.loc[i_oob, 0] += estimator.predict_proba(X.loc[i_oob])[:,1]
    num_oob.loc[i_oob, 0] += 1

""" Finally, store the probabilities assigned by this approach """
results['output_bag'] = sum_oob / num_oob
```

```python
## by taking a threshold value of probability decide the positive and negative classs as 1 and 0 respectively.

y2_comp=[]
for i in range(len(results['truth'])):
    if results['output_bag'][i] > 0.40:  ####....here i tried many probability values and took best one according to performance
        y2_comp.append(1)
    else:
        y2_comp.append(0)

# Calculate accuracy
accuracy = accuracy_score(results['truth'], y2_comp)

# Calculate F1 score
f1 = f1_score(results['truth'], y2_comp)

# Calculate AUC-ROC score
auc_roc = roc_auc_score(results['truth'], y2_comp)

# Get ROC curve values
#fpr, tpr, thresholds = roc_curve(actual_label, f_pred)

print("Accuracy:", accuracy)
print("F1 score:", f1)
print("AUC-ROC score:", auc_roc)
        
```

2.Use the positive and negative data points to train a standard classifier and apply it to the remaining unlabeled points.
